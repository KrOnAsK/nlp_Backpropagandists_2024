```python
# # Llama Text Classification Setup
# First, let's install the required dependencies

# %%
# Install required packages
%pip install sentencepiece
%pip install transformers
%pip install accelerate
%pip install bitsandbytes  # For efficient 4-bit and 8-bit model loading
%pip install scipy  # Required for some transformers operations

# %% [markdown]
# ## Check Dependencies and CUDA

# %%
import sys
import torch
import logging
from importlib.util import find_spec

def check_dependencies():
    """Check if all required dependencies are installed"""
    required_packages = {
        'sentencepiece': 'SentencePiece is required for Llama tokenization',
        'transformers': 'Transformers is required for the Llama model',
        'torch': 'PyTorch is required for model operations',
        'accelerate': 'Accelerate is required for efficient model loading',
        'bitsandbytes': 'BitsAndBytes is required for 4/8-bit quantization',
        'wandb': 'Weights & Biases is required for experiment tracking'
    }
    
    missing_packages = []
    for package, message in required_packages.items():
        if find_spec(package) is None:
            missing_packages.append(f"{message} (missing: {package})")
    
    if missing_packages:
        print("Missing required packages:")
        for msg in missing_packages:
            print(f"- {msg}")
        return False
    
    print("All required packages are installed!")
    return True

# Check CUDA and dependencies
print("Checking dependencies and CUDA availability...")
print(f"Python version: {sys.version}")
print(f"PyTorch version: {torch.__version__}")
print(f"CUDA available: {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"CUDA version: {torch.version.cuda}")
    print(f"GPU device: {torch.cuda.get_device_name(0)}")
    print(f"GPU memory: {torch.cuda.get_device_properties(0).total_memory / 1e9:.2f} GB")

# Check all dependencies
dependencies_ok = check_dependencies()
if not dependencies_ok:
    raise ImportError("Please install missing packages before continuing")

# %% [markdown]
# ## Import Required Libraries

# %%
import os
import pandas as pd
import torch
import wandb
from transformers import LlamaTokenizer, LlamaForSequenceClassification, TrainingArguments, Trainer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, confusion_matrix
from datetime import datetime
import logging
import json
import numpy as np
from tqdm.notebook import tqdm

# Configure logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s'
)
logger = logging.getLogger(__name__)

print("Setup complete! You can now proceed with the training code.")
```
